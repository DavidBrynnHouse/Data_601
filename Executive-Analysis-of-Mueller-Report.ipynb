# Analysis of the Mueller Report
This notebook contains data and analysis for the Special Council Robert Mueller's Report on the Investigation Into Russian Interference in the 2016 Presidential Election. The full redacted text was used in this analysis.

---

## The Mueller Report
The Mueller Report was made public on April 18, 2019 by Attorny General William Barr. This csv version of the document was retrieved from https://www.kaggle.com/paultimothymooney/mueller-report

---

## Goal
The goal of this report was to analyze the different types of redactions within the report to determine which type was the most prevalent. Different types include: "Harm To Ongoing Matter", "Personal Privacy", and "Investigative Technique". Abbreviated as "HOM", "PP", and "IT" respectively. Unfortunatly PP and Personal Privacy do not appear in the csv, most likely because they are green on black background in the pdf so thus are not recognized by the conversion algorithm. I also wanted to get a word count of the most used words in the report to better understand what the investigators looked at.

---

## Methods
Using the find and replace methods within excel the symbols and punctuations were removed from the initial csv. In python I created two functions, one which cleaned the document so that each word became its own column and saved that to a csv, and another which counted the occurences of redactions. The function that separated each word also filtered [stopwords](https://medium.com/@tusharsri/remove-add-stop-words-7e2994c19c67) and any words under three characters of length were removed to make the data more relevent. The 100 most common words were then sorted and analyzed.




## Analysis of Use of Redactions

---

```python
'''
David House
9/11/2020

'''
import pandas as pd
redactions_df = pd.read_csv("Data/redactions.csv")
redactions_df = redactions_df.rename(index={0:'HOM',1:'IT'})
redactions_df.plot.pie(y='Count')
```

I chose to use a pie chart because it clearly shows the percentage of uses of different types of redactions.

From this char we can see that most of the redactions were due to 'Harm to Ongoing Matter'. This could mean that investigations are still taking place that could reveal more light on this report. The minority of redactions were due to 'Investigative Techniques' which in all liklyhood will never be know to the public. Unfortunatly the third category 'Personal Privacy' was unable to be detcted by the conversion algorithm, though it is safe to say that those facts would not be known to the public ever either.

## Analysis of Word Cound

```python
cleaned_mueller_df = pd.read_csv("Data/cleaned_mueller.csv")
cleaned_mueller_df.plot.barh(y='Counts', x='Words',width=.85,figsize=(10,5))
```
