# Analysis of the Mueller Report
This notebook contains data and analysis for the Special Council Robert Mueller's Report on the Investigation Into Russian Interference in the 2016 Presidential Election.

---

## The Mueller Report
The Mueller Report was made public on April 18, 2019 by Attorny General William Barr. This csv version of the document was retrieved from https://www.kaggle.com/paultimothymooney/mueller-report

---

## Goal
The goal of this report was to analyze the word count in this document to determine the major focus of this report.

---

## Methods
Using the find and replace methods within excel the symbols and punctuations were removed from the initial csv. Then using Pandas each word in the csv was loaded into a `DataFrame`. Using this `DataFrame` the [stopwords](https://medium.com/@tusharsri/remove-add-stop-words-7e2994c19c67) and any words under three characters of length were removed to make the data more relevent. The 100 most common words were then sorted and analyzed.




```python
'''
David House
9/11/2020

'''

import json
import pandas as pd
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

# List of words such as 'the' 'and' 'of' which are not relavant to analysis
stop_words = set(stopwords.words('english'))

#loads in data retrieved from https://www.kaggle.com/paultimothymooney/mueller-report
mueller_csv = pd.read_csv("../Data/mueller_report.csv")
words_csv = pd.read_csv("../Data/words.csv")

#dict for containing the full text of report
dict_of_full_text = {}

#fills dict with relevent words from data set
for index, row in mueller_csv.iterrows():
    if type(row['text']) == str:
        for word in row['text'].split():
            if not word in stop_words and len(word) > 3:
                if word in dict_of_full_text:
                    dict_of_full_text[word] = dict_of_full_text[word] + 1
                else:
                    dict_of_full_text[word] = 1

# Orders data and displays 100 most used words
list_of_words = list(dict_of_full_text)
list_of_counts = list(dict_of_full_text.values())
final_df = pd.DataFrame({'Words':list_of_words,'Counts':list_of_counts})
final_df.sort_values(by='Counts', ascending=False, inplace=True)
final_df.head(100).plot.barh(y='Counts', x='Words',width=1,figsize=(10,20))
```
